```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
#turn column "who" into integers 
who_num = {'man':1, 'woman':2, 'child':3}
titanic['who'] = titanic['who'].apply(lambda x: who_num[x])
```

```python
#turn column "sex" into integers
sex_num = {'male':1, 'female':2}
titanic['sex'] = titanic['sex'].apply(lambda x: sex_num[x])
```
